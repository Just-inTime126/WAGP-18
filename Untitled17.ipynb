```python
import json
import os
import torch
from transformers import AutoTokenizer, AutoModel
import numpy as np
import Levenshtein
from openai import OpenAI
from google.colab import userdata
import openai
import markdown
from IPython.display import display, Markdown, HTML
import re
import time

# Programm zum Hinzufügen und Abrufen von Aufgaben
def text_eintragen():
    dateiname = "texte.json"

    # Überprüfen, ob die Datei bereits existiert
    if os.path.exists(dateiname):
        with open(dateiname, "r", encoding="utf-8") as datei:
            daten = json.load(datei)  # Daten aus der Datei laden
    else:
        daten = {}  # Leeres Dictionary, falls die Datei nicht existiert

    while True:
        aufgaben_daten = {}  # Dictionary für die Daten einer Aufgabe

        # Eingabe des eigentlichen Aufgabentextes ("Inhalt")
        inhalt = input("Aufgabentext für diese Aufgabe eingeben: ")
        aufgaben_daten["inhalt"] = inhalt  # Inhalt speichern

        # ID basierend auf der Anzahl der vorhandenen Aufgaben
        id = len(daten) + 1
        aufgaben_daten["id"] = id  # ID speichern

        # Attribute für die Aufgabe eingeben
        task_id = input(f"task-id für diese Aufgabe eingeben: ")
        aufgaben_daten["task-id"] = task_id

        # Überprüfen, ob die task-id bereits existiert
        if any(aufgabe.get("task-id") == task_id for aufgabe in daten.values()):
            entscheidung = input(f"Eine Aufgabe mit der task-id '{task_id}' existiert bereits. "
                                 f"Möchten Sie die vorhandene Aufgabe behalten (b) oder ersetzen (e)? ")
            if entscheidung.lower() == "b":
                continue  # Aufgabe überspringen und mit der nächsten fortfahren
            else:
                # Vorhandene Aufgabe entfernen (falls gewünscht)
                for vorhandene_id, vorhandene_aufgabe in list(daten.items()):
                    if vorhandene_aufgabe.get("task-id") == task_id:
                        del daten[vorhandene_id]
                        break

        aufgaben_daten["Schwierigkeit"] = input(f"Schwierigkeit für diese Aufgabe eingeben: ")
        aufgaben_daten["Typ"] = input(f"Typ für diese Aufgabe eingeben: ")

        # Daten der Aufgabe zu den Hauptdaten hinzufügen
        daten[id] = aufgaben_daten  # Aufgabe mit ID als Key speichern

        # Abfrage nach Fortsetzung oder Beendigung
        weiter = input("Weitere Aufgaben eingeben? (ja/nein): ")
        if weiter.lower() != "ja":
            break  # Beendet die äußere Schleife

    # Daten in die Datei speichern
    with open(dateiname, "w", encoding="utf-8") as datei:
        json.dump(daten, datei, indent=4, ensure_ascii=False)

    print("Daten in 'texte.json' gespeichert.")

def text_suchen():
    try:
        with open("texte.json", "r", encoding="utf-8") as datei:
            daten = json.load(datei)

        # Suche nach Aufgaben mit bestimmtem Attribut
        attribut = input("Attribut auswählen (task-id, Schwierigkeit, Typ): ")
        wert = input(f"Wert für das Attribut '{attribut}' eingeben: ")

        for id, aufgabe in daten.items():
            if aufgabe.get(attribut) == wert:  # Überprüfen, ob das Attribut vorhanden ist und den Wert hat
                print(f"Aufgabe {id}: {aufgabe['inhalt']}")

    except FileNotFoundError:
        print("Datei 'texte.json' nicht gefunden. Bitte zuerst Daten eintragen.")

# Programm zum Hinzufügen von Fragen zu einer Fragendatenbank

def fragen_hinzufuegen_mit_schlagwoertern():
    """Fügt Fragen mit Schlagwörtern hinzu und verwaltet die Schlagwörter.
    Überschreibt bestehende Fragen nach Bestätigung."""

    try:
        with open('fragen_antworten.json', 'r') as f:
            fragen_antworten = json.load(f)
    except FileNotFoundError:
        fragen_antworten = {}

    try:
        with open('schlagwoerter.json', 'r') as f:
            alle_schlagwoerter = json.load(f)
    except FileNotFoundError:
        alle_schlagwoerter = []

    while True:
        neue_frage = input("Gib die neue Frage ein (oder 'exit' zum Beenden): ")
        if neue_frage.lower() == 'exit':
            break

        # Überprüfen, ob die Frage bereits existiert
        if neue_frage in fragen_antworten:
            print(f"""Die folgende Frage existiert bereits: '{neue_frage}'
Mit der Antwort: '{fragen_antworten[neue_frage]['antwort']}'
Und den Schlagwörtern: '{fragen_antworten[neue_frage]['schlagwoerter']}'""")
            ueberschreiben = input("Soll sie überschrieben werden? (Ja/Nein): ")
            if ueberschreiben.lower() != 'ja':
                continue  # Frage überspringen, wenn nicht überschrieben werden soll

        neue_antwort = input("Gib die Antwort auf die Frage ein: ")

        # Schlagwörter abfragen und verarbeiten
        schlagwoerter_str = input("Gib Schlagwörter für die Frage ein, getrennt durch Kommas: ")
        neue_schlagwoerter = [wort.strip() for wort in schlagwoerter_str.split(",")]

        # Schlagwörter überprüfen und hinzufügen
        frage_schlagwoerter = []
        for schlagwort in neue_schlagwoerter:
            if schlagwort in alle_schlagwoerter:
                frage_schlagwoerter.append(schlagwort)
            else:
                alle_schlagwoerter.append(schlagwort)
                frage_schlagwoerter.append(schlagwort)
                print(f"Neues Schlagwort '{schlagwort}' hinzugefügt.")

        # Frage, Antwort und Schlagwörter zur Datenbank hinzufügen
        fragen_antworten[neue_frage] = {
            "antwort": neue_antwort,
            "schlagwoerter": frage_schlagwoerter
        }

        print(f"Die Frage '{neue_frage}' wurde mit der Antwort '{neue_antwort}' und den Schlagwörtern '{frage_schlagwoerter}' zur Datenbank hinzugefügt.")

    # Aktualisierte Datenbanken speichern
    with open('fragen_antworten.json', 'w') as f:
        json.dump(fragen_antworten, f, indent=4)
    with open('schlagwoerter.json', 'w') as f:
        json.dump(alle_schlagwoerter, f, indent=4)

    print("Die Datenbanken wurden aktualisiert.")

# Programm zum Tokenisieren der Fragen

def berechne_vektor(text):
    """Berechnet den Vektor für einen Text."""
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state[:, 0, :]  # CLS-Token Embedding
    return embeddings[0].detach().numpy().tolist()  # Vektor als Liste

def Fragen_tokenisieren():
  # Lade das SciBERT-Modell und den Tokenizer
  tokenizer = AutoTokenizer.from_pretrained("allenai/scibert_scivocab_uncased")
  model = AutoModel.from_pretrained("allenai/scibert_scivocab_uncased")

  # Lade die Fragen und Antworten aus der JSON-Datei
  with open('fragen_antworten.json', 'r') as f:
      fragen_antworten = json.load(f)

  # Berechne die Vektoren für alle Fragen und speichere sie in einem Dictionary
  fragen_vektoren = {}
  schlagwort_vektoren = {}  # Dictionary für Schlagwortvektoren

  for frage, data in fragen_antworten.items():
      fragen_vektoren[frage] = berechne_vektor(frage)

      # Schlagwörter extrahieren und Vektoren berechnen
      for schlagwort in data.get("schlagwoerter", []):
          if schlagwort not in schlagwort_vektoren:  # Vermeide doppelte Berechnung
              schlagwort_vektoren[schlagwort] = berechne_vektor(schlagwort)

  # Speichere die Vektoren in separaten JSON-Dateien
  with open('fragen_vektoren.json', 'w') as f:
      json.dump(fragen_vektoren, f)

  with open('schlagwort_vektoren.json', 'w') as f:
      json.dump(schlagwort_vektoren, f)

  print("Vektoren wurden berechnet und in 'fragen_vektoren.json' und 'schlagwort_vektoren.json' gespeichert.")

# passende Antworten zu den gestellten Fragen finden
def berechne_vektor(frage):
    """Berechnet den Vektor für eine Frage."""
    inputs = tokenizer(frage, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state[:, 0, :]  # CLS-Token Embedding
    return embeddings[0].detach().numpy()  # Vektor als NumPy-Array

def finde_aehnlichste_frage(frage, schwellenwert=0.9):
    """Findet die ähnlichste Frage."""
    frage_vektor = berechne_vektor(frage)

    beste_aehnlichkeit = 0
    aehnlichste_frage = None

    for db_frage, db_vektor in fragen_vektoren.items():
        aehnlichkeit = np.dot(frage_vektor, np.array(db_vektor)) / (np.linalg.norm(frage_vektor) * np.linalg.norm(np.array(db_vektor)))

        if aehnlichkeit > beste_aehnlichkeit:
            beste_aehnlichkeit = aehnlichkeit
            aehnlichste_frage = db_frage

    if beste_aehnlichkeit >= schwellenwert:
        return aehnlichste_frage
    else:
        return None

def ueberpruefe_schlagwoerter(frage, aehnlichste_frage, schwellenwert_levenshtein=0.8):
    """Überprüft die Schlagwörter mit Fehlertoleranz."""
    schlagwoerter = fragen_antworten.get(aehnlichste_frage, {}).get("schlagwoerter", [])

    for schlagwort in schlagwoerter:
        schlagwort_gefunden = False
        for wort in frage.lower().split():
            # Levenshtein-Distanz berechnen
            distanz = Levenshtein.distance(schlagwort.lower(), wort)
            aehnlichkeit = 1 - (distanz / max(len(schlagwort), len(wort)))

            if aehnlichkeit >= schwellenwert_levenshtein:
                schlagwort_gefunden = True
                break  # Schlagwort gefunden, weiter zum nächsten Schlagwort

        if not schlagwort_gefunden:
            return False  # Mindestens ein Schlagwort nicht gefunden

    return True  # Alle Schlagwörter gefunden

def Frage_beantworten():
  # Lade das SciBERT-Modell und den Tokenizer
  tokenizer = AutoTokenizer.from_pretrained("allenai/scibert_scivocab_uncased")
  model = AutoModel.from_pretrained("allenai/scibert_scivocab_uncased")

  # Lade die Fragenvektoren aus der JSON-Datei
  with open('fragen_vektoren.json', 'r') as f:
      fragen_vektoren = json.load(f)

  # Lade die Fragen und Antworten aus der JSON-Datei
  with open('fragen_antworten.json', 'r') as f:
      fragen_antworten = json.load(f)

  while True:
      frage = input("Stelle eine Frage (oder 'exit' zum Beenden): ")
      if frage.lower() == 'exit':
          break

      aehnlichste_frage = finde_aehnlichste_frage(frage)

      if aehnlichste_frage:
          if ueberpruefe_schlagwoerter(frage, aehnlichste_frage):
              antwort = fragen_antworten.get(aehnlichste_frage, {}).get("antwort", "Ich habe keine passende Antwort gefunden.")
              print(antwort)
          else:
              print(f"Meintest du die Frage: '{aehnlichste_frage}'?")
              feedback = input("Ja/Nein: ")
              if feedback.lower() == "ja":
                  antwort = fragen_antworten.get(aehnlichste_frage, {}).get("antwort", "Ich habe keine passende Antwort gefunden.")
                  print(antwort)
              else:
                  print("Ich habe keine passende Antwort gefunden.")
                  KI_Frage = input("Möchtest du die Antwort von einer KI generieren lassen? (Ja/Nein): ")
                  if KI_Frage.lower() == "ja":
                    Fragen_Antwort_Bot(frage)

      else:
          print("Ich habe keine passende Antwort gefunden.")
          KI_Frage = input("Möchtest du die Antwort von einer KI generieren lassen? (Ja/Nein): ")
          if KI_Frage.lower() == "ja":
            Fragen_Antwort_Bot(frage)

# Programm zum Beantworten von Fragen durch KI

def Fragen_Antwort_Bot(Frage):
  # Setze deinen OpenAI API-Key
  api_key = userdata.get("Open-AI")
  client = OpenAI(api_key = api_key)
  rolle = "Du bist ein KI-Tutor, der Studierenden der Technischen Hochschule Köln am Campus Gummersbach im Mechanikmodul des ingenieurwissenschaftlichen Grundstudiums helfen soll, indem du ihnen Fragen beantwortest. Wir bewegen uns hier nur in zwei Dimensionen."
  messages = [{"role": "developer", "content": rolle},
              {"role": "user", "content": Frage}]

  try:
    response = client.chat.completions.create(
        model="gpt-4o",
        messages= messages
    )
  except openai.error.RateLimitError:
    print("Das Anfragenlimit ist erreicht. Es muss für 60 Sekunden gewartet werden.")
    time.sleep(60)
    response = client.chat.completions.create(
        model="gpt-4o",
        messages= messages
    )

  ausgabe = response.choices[0].message.content

  print(ausgabe)

# KI zum Aufgabenerklären

def aufgabe_laden():
    try:
        with open("texte.json", "r", encoding="utf-8") as datei:
            daten = json.load(datei)

        task_id = input("Bitte geben Sie die Aufgabenkennung ein (z.B. 1.2): ")

        # Aufgabe anhand der task-id finden
        for id, aufgabe in daten.items():
            if aufgabe.get("task-id") == task_id:
                aufgaben_inhalt = aufgabe["inhalt"]
                return aufgaben_inhalt

        print(f"Keine Aufgabe mit der task-id '{task_id}' gefunden.")
        return None  # Rückgabewert None, falls keine Aufgabe gefunden wurde

    except FileNotFoundError:
        print("Datei 'texte.json' nicht gefunden. Bitte stellen Sie sicher, dass die Datei existiert.")
        return None

def process_task(aufgaben_text, fixed_instruction):
    """Verarbeitet die Aufgabe mit der KI."""

    api_key = userdata.get('Open-AI')
    client = OpenAI(api_key=api_key)
    combined_instruction = f"{fixed_instruction} Aufgabenstellung: {aufgaben_text}"

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "developer", "content": "Du bist ein KI-Tutor, der Studierenden dabei helfen soll, ihre Aufgaben im Mechanik-Modul des ingenieurwissenschaftlichen Grundstudiums an der TH Köln zu lösen."},
                {"role": "user", "content": combined_instruction}
            ],
            max_tokens=5000
        )
    except openai.error.RateLimitError:
        print("Das Anfragenlimit ist erreicht. Es muss für 60 Sekunden gewartet werden.")
        time.sleep(60)
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "developer", "content": "Du bist ein KI-Tutor, der Studierenden dabei helfen soll, ihre Aufgaben im Mechanik-Modul des ingenieurwissenschaftlichen Grundstudiums an der TH Köln zu lösen."},
                {"role": "user", "content": combined_instruction}
            ],
            max_tokens=5000
        )

    text_response = response.choices[0].message.content

    return text_response


def Latex_filter(text):
  output = re.sub(r"\\\(", r"$", text)
  output = re.sub(r"\\\)", r"$", output)
  output = re.sub(r"\\\[", r"$", output)
  output = re.sub(r"\\\]", r"$", output)
  return output

def Aufgaben_erklären():
  # Feste Anweisung im Quellcode
  fixed_instruction = "Du bekommst im folgenden Text eine Aufgabenstellung. Löse diese Aufgabe bitte möglichst ausführlich unter Anführen aller nötigen Lösungsschritte. Formatiere das Ergebnis so, dass es von Markdown verarbeitet werden kann."

  # Aufgabentext aus der JSON-Datei laden
  aufgaben_text = aufgabe_laden()

  # process_task aufrufen und Antwort erhalten, wenn aufgaben_text erfolgreich geladen wurde
  antwort = process_task(aufgaben_text, fixed_instruction)

  ausgabe = Latex_filter(antwort)

  display(Markdown(ausgabe))

# Tipps für das Aufgabenlösen

def Tipps_Aufgabenlösen():
  # Setze deinen OpenAI API-Key
  api_key = userdata.get("Open-AI")
  client = OpenAI(api_key = api_key)
  rolle = "Du bist ein KI-Tutor, der Studierenden der Technischen Hochschule Köln am Campus Gummersbach im Mechanikmodul des ingenieurwissenschaftlichen Grundstudiums helfen soll, indem du ihnen Tipps zum Lösen von Aufgaben gibst. Berechnungen werden dabei jeweils nur in zwei Dimensionen durchgeführt."
  while True:
    Frage = input("Was möchtest Du wissen? (oder \"exit\" zum Beenden):")
    if Frage.lower() == "exit":
      break
    messages = [{"role": "developer", "content": rolle},
                {"role": "user", "content": Frage}]

    try:
      response = client.chat.completions.create(
          model="gpt-4o-mini",
          messages= messages
      )
    except openai.error.RateLimitError:
      print("Das Anfragenlimit ist erreicht, es muss für 60 Sekunden gewartet werden.")
      time.sleep(60)
      response = client.chat.completions.create(
          model="gpt-4o-mini",
          messages= messages
      )

    ausgabe = response.choices[0].message.content

    Markdown_Ausgabe = Latex_filter(ausgabe)

    display(Markdown(Markdown_Ausgabe))

# Tutand

def themen_datei_erweitern():
  dateiname = "themen.json"
  try:
      with open(dateiname, "r") as datei:
          themen = json.load(datei)  # Themen aus der Datei laden
  except FileNotFoundError:
      themen = {}  # Leeres Dictionary erstellen, falls Datei nicht existiert

  while True:
      thema = input("Gib ein Thema ein (oder 'exit' zum Beenden): ")
      if thema.lower() == "exit":
          break

      # Prüfung, ob Thema bereits existiert
      if thema in themen.values():
          print("Thema existiert bereits. Bitte gib ein anderes Thema ein.")
          continue  # Schleife fortsetzen

      # Automatische Nummerierung der Themen
      zahl = len(themen) + 1
      themen[str(zahl)] = thema

  with open(dateiname, "w") as datei:
      json.dump(themen, datei, indent=4)

  print("Die Datei 'themen.json' wurde aktualisiert.")

def themen_laden():
  with open("themen.json", "r") as datei:
    themen = json.load(datei)
  return themen

def themen_ausgeben(themen):
  for zahl, thema in themen.items():
    print(f"  {zahl}: {thema}")

def thema_auswaehlen(themen):
  while True:
    eingabe = input("Bitte wähle ein Thema aus (oder \"exit\" zum Beenden): ")
    if eingabe in themen:
      return themen[eingabe]
      break
    elif eingabe.lower() == "exit":
      return "exit"
    else:
      print("Ungültige Eingabe. Bitte versuche es erneut.")

def KI_Tutand_Abfrage(fixed_instruction, rolle, Thema, Abfrage):
  api_key = userdata.get('Open-AI')
  client = OpenAI(api_key=api_key)
  combined_instruction = f"{fixed_instruction} {Thema} Du hast bereits folgende Abfragen gestellt, frage daher etwas anderes ab: {Abfrage}"

  try:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "developer", "content": rolle},
            {"role": "user", "content": combined_instruction}
        ],
        max_tokens=500
    )
  except openai.error.RateLimitError:
    print("Das Anfragenlimit ist erreicht. Es muss für 60 Sekunden gewartet werden.")
    time.sleep(60)
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "developer", "content": rolle},
            {"role": "user", "content": combined_instruction}
        ],
        max_tokens=500
    )

  text_response = response.choices[0].message.content
  Abfrage = text_response
  display(Markdown(text_response))
  return text_response

def KI_Tutor_Bewertung(fixed_instruction, rolle, Thema, Antwort, Abfrage):
  api_key = userdata.get('Open-AI')
  client = OpenAI(api_key=api_key)
  combined_instruction = f"{fixed_instruction} Du hast folgende Abfrage gestellt: {Abfrage} Es wurde folgende Antwort gegeben: {Antwort}"

  try:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "developer", "content": rolle},
            {"role": "user", "content": combined_instruction}
        ],
        max_tokens=500
    )
  except openai.error.RateLimitError:
    print("Das Anfragenlimit ist erreicht. Es muss für 60 Sekunden gewartet werden.")
    time.sleep(60)
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "developer", "content": rolle},
            {"role": "user", "content": combined_instruction}
        ],
        max_tokens=2000
    )

  text_response = response.choices[0].message.content
  Abfrage = text_response
  display(Markdown(text_response))
  return text_response

def score_aktualisieren(neue_zahl):
  """
  Aktualisiert eine Zahl in der JSON-Datei und hängt neue Zahlen an.

  Args:
      dateiname: Name der JSON-Datei.
      index: Index der zu aktualisierenden Zahl in der Liste.
      neuer_wert: Der neue Wert für die Zahl.
      neue_zahlen: Eine Liste neuer Zahlen, die angehängt werden sollen.
  """
  dateiname = "score.json"
  try:
    with open(dateiname, "r") as datei:
      zahlen = json.load(datei)  # Aktuelle Zahlen aus der Datei laden
  except FileNotFoundError:
    zahlen = [0]
    with open(dateiname, "w") as datei:
      json.dump(zahlen, datei)
  zahlen.append(neue_zahl)  # Neue Zahlen anhängen
  i = 0
  score = 0
  for element in zahlen[1:]:
    #i += 1
    score = score + int(element)

  score = score/(len(zahlen) - 1)
  zahlen[0] = score
  return score
  with open(dateiname, "w") as datei:
    json.dump(zahlen, datei)  # Aktualisierte Zahlen speichern

def KI_Tutand():
  while True:
    themen = themen_laden()
    print("Es stehen folgende Themen zur Auswahl:")
    themen_ausgeben(themen)
    ausgewaehltes_thema = thema_auswaehlen(themen)
    if ausgewaehltes_thema == "exit":
      break
    else:
      print(f"Ausgewähltes Thema: {ausgewaehltes_thema}")
    fixed_instruction_Abfrage = """ Du sollst das Wissen eines Studierenden an der Technischen Hochschule Köln am Campus Gummersbach evaluieren.
                          Dazu sollst du Wissen abfragen, indem du dich als Student ausgibst, der etwas von einem Studierenden wissen möchte.
                          Nachdem der Studierende seine Antwort gegeben hat, sollst du diese bewerten, indem du sie von 0 \(das ist der schlechteste Wert\) bis 10 \(das ist der schlechteste Wert\) bewertest.
                          Dabei sollst du aber Sätze wie \"Klar, kann ich machen!\" auslassen.
                          Sätze, in denen du dich als Kommilitone ausgibst, z.B. \"Hi, ich bin auch Student hier. Kannst du mir bitte folgendes erklären?\" oder \"Hallo, ich bin einer deiner Kommilitonen, kannst du mir helfen?\" kannst du gerne verwenden.
                          Gib dabei gerne auch zu verstehen, dass du ein Thema nicht verstehst.
    """
    rolle = """Du bist ein KI-Tutor, der speziell dafür gedacht ist, das Wissen von Studierenden abzufragen und zu bewerten."""
    Abfrage_1 = "Du hast noch keine Abfrage gestellt. Stelle daher deine erste Anfrage."
    fixed_instruction_Bewertung = """Du sollst das Wissen von Studierenden bewerten, indem du die von ihnen gegebene Antwort bewertest.
                                    Dabei sollst du die gegebene Antwort in einer Skala von 0 als schlechtester Wert bis 10 als bester Wert bewerten.
                                    Die Skala soll in folgender Form angegeben werden: Wert, der erreicht wurde/Wert, der hätte erreicht werden können
                                    Gib zu den gegebenen Antworten bitte Feedback, füge die Skala aber bitte ohne Kommentar (z.B. \"Die Bewertung lautet:\" oder \"Wert, der erreicht wurde\" oder \"Deine Punktzahl:\") mit einer Leerzeile an das Ende deiner Bewertung an.
                                    """
    Thema = f"Du sollt den Studierenden über folgendes Thema abfragen: {ausgewaehltes_thema}"
    Abfrage_neu = KI_Tutand_Abfrage(fixed_instruction_Abfrage, rolle, Thema, Abfrage_1)
    while True:
      Antwort_Studi = input("Bitte gib deine Antwort ein (oder \"exit\" zum Beenden): ")
      if Antwort_Studi.lower() == "exit":
        break
      Bewertung = KI_Tutor_Bewertung(fixed_instruction_Bewertung, rolle, Thema, Antwort_Studi, Abfrage_neu)
      try:
        Abfrage = f"{Abfrage} UND {Abfrage_neu}"
      except NameError:
        Abfrage = Abfrage_neu
      Teile = Bewertung.split("\n\n")
      Skala = Teile[-1]
      Skala_split = Skala.split("/")
      erreichte_Punkte = int(Skala_split[-2])
      maximale_Punkte = int(Skala_split[-1])
      erreichter_score = score_aktualisieren(erreichte_Punkte)
      gerundeter_score = round(erreichter_score, 1)
      print(f"Dein aktueller Score beträgt: {erreichter_score}/10")
      Abfrage_neu = KI_Tutand_Abfrage(fixed_instruction_Abfrage, rolle, Thema, Abfrage_1)

def Tutand_Programm():
  while True:
    print("""Bitte wählen Sie eine der folgenden Optionen:
    1: Themen hinzufügen
    2: Themen einsehen
    3: Tutand nutzen
    4: zurück
    """)
    Auswahl_Tutand = input("Ihre Auswahl: ")
    if Auswahl_Tutand == "1":
      themen_datei_erweitern()
    elif Auswahl_Tutand == "2":
      themen = themen_laden()
      themen_ausgeben(themen)
      Themen_Einsehen_beenden = input("Zum Beenden bitte \"exit\" eingeben: ")
      if Themen_Einsehen_beenden.lower() == "exit":
        pass
    elif Auswahl_Tutand == "3":
      KI_Tutand()
    elif Auswahl_Tutand == "4":
      break
    else:
      print("Ungültige Eingabe. Bitte wählen Sie eine der Optionen aus.")

# Hauptprogramm
while True:
  print("Willkommen bei unserer KI-Lernunterstützung für das Mechanik-Modul!")
  print("""Bitte wählen Sie eine der folgenden Optionen:
  1: Aufgabendatenbank
  2: Frage-/Antwortdatenbank
  3: Tutor
  4: Beenden
      """)
  Auswahl = input("Ihre Auswahl: ")

  if Auswahl == "1":
    while True:
      print("""Bitte wählen Sie eine der folgenden Optionen:
  1: Aufgabendatenbank verändern
  2: Aufgabendatenbank einsehen
  3: zurück
          """)
      Auswahl_Aufgabendatenbank = input("Ihre Auswahl: ")

      if Auswahl_Aufgabendatenbank == "1":
        text_eintragen()
      elif Auswahl_Aufgabendatenbank == "2":
        text_suchen()
      elif Auswahl_Aufgabendatenbank == "3":
        break
      else:
        print("Ungültige Auswahl. Bitte versuchen Sie es erneut.")

  if Auswahl == "2":
      while True:
        print("""Bitte wählen Sie eine der folgenden Optionen:
  1: Frage-/Antwortdatenbank verändern
  2: Frage-/Antwortdatenbank einsehen
  3: Tokenisierung anstoßen
  4: zurück
          """)
        Auswahl_FrageAntwortdatenbank = input("Ihre Auswahl: ")

        if Auswahl_FrageAntwortdatenbank == "1":
          fragen_hinzufuegen_mit_schlagwoertern()
        elif Auswahl_FrageAntwortdatenbank == "2":
          pass
        elif Auswahl_FrageAntwortdatenbank == "3":
          Fragen_Tokenisieren()
        elif Auswahl_FrageAntwortdatenbank == "4":
          break
        else:
          print("Ungültige Auswahl. Bitte versuchen Sie es erneut.")

  if Auswahl == "3":
      while True:
        print("""Bitte wählen Sie eine der folgenden Optionen:
  1: Frage stellen
  2: Lösungsweg erklären lassen
  3: Tipps für das Aufgabenlösen
  4: KI-Tutand aufrufen
  5: zurück
          """)
        Auswahl_Tutor= input("Ihre Auswahl: ")

        if Auswahl_Tutor == "1":
          Frage_beantworten()
        elif Auswahl_Tutor == "2":
          Aufgaben_erklären()
        elif Auswahl_Tutor == "3":
          Tipps_Aufgabenlösen()
        elif Auswahl_Tutor == "4":
          Tutand_Programm()
        elif Auswahl_Tutor == "5":
          break
        else:
          print("Ungültige Auswahl. Bitte versuchen Sie es erneut.")

  if Auswahl == "4":
    break
  else:
    print("Ungültige Auswahl. Bitte versuchen Sie es erneut.")

#text_eintragen()
#text_suchen()
```
